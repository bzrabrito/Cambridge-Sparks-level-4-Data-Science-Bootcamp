### Q1) Write a for-loop that prints out the numbers from 1 to 15 inclusively.

```python
#add your code below

for i in range(1, 16):
    print('Iteration:', i)
    
```

### Q3) Iterate through the list of animals and print the capitalized version of each.

Hint: there is a function `.upper()` which can be applied to strings.

```python
animals = ['duck', 'rat', 'boar', 'elefant', 'mammoth', 'gazelle']
```

```python
#add your code below

for animal in animals:
    print(animal.upper())
    
```

### Q4) Iterate through the list of animals from the previous exercise. Capitalize the first letter and append the modified animals to a new list.

Hint: You have to initialise the list to which you want to append first, just assign an empty list (one without elements) to a variable name.

```python
#add your code below

cap_animals = []
for animal in animals:
    cap_animal = animal[0].upper() + animal[1:]
    cap_animals.append(cap_animal)
print(cap_animals)

```

### Q5) Use a while loop to sum the integers from 1 to 100 inclusively.

```python
#add your code below

i = 0
res = 0
while i < 101:
    res = res+i
    i += 1
res

```

### Q6) Write a while loop which prints 'inside loop' each time it runs through the loop. Don't forget to limit the number of times you want to run through the loop.

```python
#add your code below

i = 0
while i < 4:
    print("inside loop")
    i += 1
print("outside loop")

```

### Q7) Write a while loop that removes in each traversal the last element from my_list and stops as soon as my_list is empty.

```python
#add your code below

my_list = list(range(20))

while len(my_list):
    print(my_list)
    my_list.pop()

my_list

```

```python
#add your code below

my_list = list(range(20))

while len(my_list):
    print(my_list)
    my_list.remove(my_list[-1])

my_list

```
