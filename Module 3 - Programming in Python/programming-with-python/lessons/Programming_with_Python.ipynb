# Boolean operators

We can use the following **operators** within **statements** containing Boolean (`True` or `False`) values:  

`and`: **both elements** must be `True` to produce a `True` result  
`or`: **at least one element** must be `True` to produce a `True` result    
`not`: **negates** the given result

We also have the following operators for **comparing values**.  
The following can be used with objects of **any data type**:  

`==` equal to  
`!=` **not** equal to  

The following only work with **numeric** data types:  

`>` greater than  
`<` less than

`>=` greater than or equal to  
`<=` less than or equal to



````python
age=25
weight=90

print(age>10 and weight>80)   #True AND True  --> compound effect is True
print(age>10 and weight>100)  #True AND False --> compound effect is False
````

````python
age=25
weight=90

print(age>10 or weight>80)     #True OR True  --> compound effect is True
print(age>10 or weight>100)    #True OR False --> compound effect is True
````

````python
age=25

print(age>10) 
print(not age>10)              #Reverse the result, returns False if the result is True
````

# Conditional expressions

Conditional expressions are used when we want the execution of parts of our code to be dependent on whether a condition evaluates to `True` or `False`.  
  
### `if` statements

````python
age = 17

if age < 18:
    print('Child')
````

### `if` statements

- the statement begins with the `if` keyword
- this is followed by a **boolean expression** (the condition)
- the line ends with a colon `:`
- the code to be exectued if the condition is met (i.e. `True`) is **indented** on the following line(s)

### `else` statements

````python
age = 17

if age < 18:
    print('Child')

else:
    print('Adult')
````

### `else` statements
- the **indentation** and use of a **colon** are the same as for the `if` statement
- no value of `age` can result in both `print()` statements being executed

### `elif` statements

````python
age = 77

if age < 18:    
    category = 'Child'

elif age >= 65:    
    category = 'Pensioner'
    
else:
    category = 'Adult'
    
print(category)
````

### `elif` statements

The `elif` keyword allows us to specify further conditions to be evaluated after an `if` statement:

Notice how the `print()` statement is after i.e. *outside* the conditional expression.

Python will evaluate each condition in turn and execute the first (and only the first) one which evaluates to `True`.

- The `elif` and `else` statements are **optional**
- We can have numerous `elif` statements



If no conditions are matched and there is no `else` statement, Python will move on and continue to evaluate the subsequent code. 

````python
category = None
age = 64

if age < 0 or age > 150:    
    print('Invalid age entered')

elif age >= 65:    
    category = 'Pensioner'

elif age >= 18:    
    category = 'Adult'
    
else:
    category = 'Child'
    
print(category)
````

- We can evaluate multiple conditions together using `or`, `and`, `not`
- The **indented code blocks** are only executed when the condition is `True`
- Any code before or after the conditional expressions is executed as normal

# Loops

A **loop** is a **sequence of instructions** that is **repeated** until a **condition** is met.

- Computers can perform repetitive operations very quickly
- 'Don't Repeat Yourself' (**DRY**) principle when writing code


Each time the code within a loop is executed is considered an **iteration**.

### Syntax

The general syntax for a **for loop** is indicated below:

```python
    for iterator_name in iterating_sequence:
        …statements…
```

This is a for loop. It has the heading starting with `for`, followed by a
variable name (count in this case), the word `in`, some sequence, and a final colon.
As with function definitions and other heading lines, the colon at the end of
the line indicates that a consistently indented block of statements follows
to complete the for loop.

### `for` loops

### `for` loops

`for` loops execute code **for each item** in a sequence (such as a list of values):

````python
names = ['Dave', 'Sara', 'Mo']

for name in names:
    print(name)
````

```python
for name in names:
    print(name)
```

- The `for` loop performs the operation *for* each value in the `names` list
- We can use any string between `for` and `in`; this simply determines how we refer to each value in the subsequent code

## simple examples

````python
for count in [1, 2, 3]:
    print(count)
    print('Yes' * count)
````

````python
for i in range(10):
    print('Hello')
````

````python
word = 'computer'
for letter in word:
    print(letter)
````

````python
teas = ['earl grey', 'darjeeling', 'english breakfast', 'lapsang souchong', 
        'oolong', 'kombucha', 'green tea']

for tea in teas:
    print(tea)
````

````python
for tea in teas:
    print(tea[0])
````

### `for` loops

````python
profiles = [('John', 46), ('Paul', 21), ('Jasmine', 26)]
ages = []

for entry in profiles:
    ages.append(entry[1])

print(ages)
print(f'Average age is {sum(ages) / len(ages)}')
    
````

- We have used a `for` loop to **iterate** over a list of tuples
- Our `profiles` list could have any number of entries

### `for` loops

````python
bill = 0
freebies = 0
prices = [4.75, 14, 4, 0, 18.50, 1.25, 5.60, 0]

for item_price in prices:
    
    if item_price == 0:
        freebies = freebies+1
        
    bill = bill+item_price    
    
print(f'Total is £{bill}, including {freebies} free items!')
````

- We can **nest** conditional statements within loops
- Variables can be modified and then used outside the loop

````python
shopping_list = ['milk', 'bread', 'yoghurt', 'pizza', 'pasta']

for item in shopping_list:
    
    if item == 'pizza':
        for person in ['Joe', 'Paula']:
            print(item + ' for ' + person)
    
    else:
        print(item)
````

### `while` loops



### Syntax

The general syntax for a **while loop** is indicated below:

```python
    while condition:
        …statements…
```

The syntax includes the following parts:
* in the first line
    * the keyword **`while`** 
    * followed by a condition which evaluates to `true` or `false`
    * followed by a colon
* an indented code block including all statements to be executed if the condition evaluates to true

### `while` loops

````python
count = 0
while count < 5:
    print(count)
    count = count + 1
````

````python
a = 0
while a < 5:
    a += 1
    print(a)
````

### `while` loops
`while` loops continually execute **while** a **conditional expression** is `True` 

````python
units = 'kg'
weight = 80
target = 72
monthly_change = -3
month = 0

while weight > target:
    month = month+1
    weight = weight+monthly_change
    print(f'Month {month}: {weight}')

print(f'It will take {month} months to reach {target}{units}.')
````

- Decrease the value of weight by 3 units, until weight is above 72

### `while` loops

- Make sure that the loop will stop eventually, i.e. that the condition will become `False` 
    - a loop which never stops is known as an **infinite loop**
    
- You are likely to use `for` loops more frequently in data analysis
    - `while` loops are more often found in **continuous processes**, for example as a means to shut off a machine if a safety alert is triggered
